#### 3. Creating visualizations

```python
#Creating a bar chart for the orders_day_of_week column
ords_prods_merge_grouped['orders_day_of_week'].value_counts().plot.bar()
```

```python
#Creating a bar chart for the orders_day_of_week column
bar = ords_prods_merge_grouped['orders_day_of_week'].value_counts().sort_index().plot.bar()
```

```python
#Adding colour to the bar chart
bar = ords_prods_merge_grouped['orders_day_of_week'].value_counts().sort_index().plot.bar(color =['purple', 'red', 'pink', 'orange', 'yellow', 'green', 'blue'])
```

```python
#Sorting the bars from highest to lowest
bar = ords_prods_merge_grouped['orders_day_of_week'].value_counts().plot.bar(color =['purple', 'red', 'pink', 'orange', 'yellow', 'green', 'blue'])
```

```python
#Exporting the bar chart
bar.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'bar_orders_dow.png'))
```

```python
#Creating a histogram of the prices column
ords_prods_merge_grouped['prices'].plot.hist(bins = 25)
```

```python
#Creating a scatterplot for the prices column
sns.scatterplot(x = 'prices', y = 'prices',data = ords_prods_merge_grouped)
```

```python
#Checking products with prices over $100
ords_prods_merge_grouped.loc[ords_prods_merge_grouped['prices'] > 100]
```

```python
#Creating a histogram for the prices column
hist = ords_prods_merge_grouped['prices'].plot.hist(bins = 25)
```

```python
#Increasing the number of bins in the histogram to 70
hist = ords_prods_merge_grouped['prices'].plot.hist(bins = 70)
```

```python
#Exporting the histogram
hist.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'hist_prices.png'))
```
