#### 3. Creating new columns

```python
#Defining a function for price_label
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
#Applying the function
df['price_range'] = df.apply(price_label, axis=1)
```

```python
#Checking the value counts for price_range
df['price_range'].value_counts(dropna = False)
```

```python
#Finding the maximum price
df['prices'].max()
```

```python
#Creating the High-range product flag in df
df.loc[df['prices'] > 15,'price_range_loc'] = 'High-range product'
```

```python
#Creating the Mid-range product flag in df
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
#Creating the Low-range product flag in df
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
#Checking the value counts for price_range_loc
df['price_range_loc'].value_counts(dropna = False)
```

```python
#Creating the High-range product flag in df_ords_prods_merged
df_ords_prods_merged.loc[df_ords_prods_merged['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
#Defining a function for busiest day
result = []

for value in df_ords_prods_merged["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
#Creating a new column for busiest days
result_2 = []

for value in df_ords_prods_merged["orders_day_of_week"]:
  if value == 0 or value == 1:
    result_2.append("Busiest days")
  elif value == 4 or value == 3:
    result_2.append("Slowest days")
  else:
    result_2.append("Regularly busy")
```

```python
#Checking result_2
result_2
```

```python
#Creating the busiest_days column
df_ords_prods_merged['busiest_days'] = result_2
```

```python
#Checking the values in the new busiest_days column
df_ords_prods_merged['busiest_days'].value_counts(dropna = False)
```

```python
#Checking the output of dataframe with new busiest days columns
df_ords_prods_merged.head()
```
