#### 3. Finding the Aggregated Mean

```python
#Finding the aggregated mean of the order_number column grouped by the department_id column
ords_prods_merge.groupby('department_id').agg({'order_number': ['mean']})
```

#### 4. Creating a Loyalty Flag

```python
#Creating the max_order column
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
#Checking if the max_order column was created
ords_prods_merge.head(15)
```

```python
#Allowing all rows to be shown with the head function
pd.options.display.max_rows = None
```

```python
#Checking the first one hundred rows of ords_prods_merge
ords_prods_merge.head(100)
```

```python
#Creating a loyalty flag for loyal customers
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
#Creating a loyalty flag for regular customers
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
#Creating a loyalty flag for new customers
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
#Checking the frequency of the loyalty flag column
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
#Checking the user_id, loyalty_flag, and order_number columns
ords_prods_merge[['user_id', 'loyalty_flag', 'order_number']].head(60)
```

```python
#Finding the basic statistics of product prices for each loyalty category
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

Question 5. The prices of products purchased by loyal customers is almost the same as the prices of products purchased by both new and regular customers.

#### 5. Creating a Spending Flag

```python
#Using the transform function to aggregate data
ords_prods_merge['avg_product_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
#Checking if the avg_product_price column was created
ords_prods_merge.head(15)
```

```python
#Checking the first one hundred rows of ords_prods_merge
ords_prods_merge.head(100)
```

```python
#Creating a spending flag for high spender
ords_prods_merge.loc[ords_prods_merge['avg_product_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
#Creating a spending flag for low spender
ords_prods_merge.loc[ords_prods_merge['avg_product_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
#Checking the frequency of the spending flag column
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
#Checking the user_id, spending_flag, and avg_product_price columns
ords_prods_merge[['user_id', 'spending_flag', 'avg_product_price']].head(50)
```

#### 6. Creating an Order Frequency Flag

```python
#Using the transform function to aggregate data
ords_prods_merge['median_days_between_orders'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
#Checking if the median_days_between_orders column was created
ords_prods_merge.head(15)
```

```python
#Checking the first one hundred rows of ords_prods_merge
ords_prods_merge.head(100)
```

```python
#Creating an order frequency flag for non-frequent customer
ords_prods_merge.loc[ords_prods_merge['median_days_between_orders'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
```

```python
#Creating an order frequency flag for regular customer
ords_prods_merge.loc[(ords_prods_merge['median_days_between_orders'] <= 20) & (ords_prods_merge['median_days_between_orders'] > 10), 'order_frequency_flag'] = 'Regular customer'
```

```python
#Creating an order frequency flag for frequent customer
ords_prods_merge.loc[ords_prods_merge['median_days_between_orders'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
#Checking the user_id, order_frequency_flag, and median_days_between_orders columns
ords_prods_merge[['user_id', 'order_frequency_flag', 'median_days_between_orders']].head(50)
```

```python
#Checking the frequency of the order_frequency_flag column
ords_prods_merge['order_frequency_flag'].value_counts(dropna = False)
```
